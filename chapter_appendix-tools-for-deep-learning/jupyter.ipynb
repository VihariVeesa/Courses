```python
# DL -- Deep Learning
# neurons -- inspried from human brain
# text, images, videos, audio , real-time data
# ANN -- Aritifical Neural network -- text based -- huge amount of text
# CNN -- convolutional NN  --- images, audio, vidoes
# RNN -- Recurrent NN -- real-time information --
# tensorflow /keras, pytorch
# word-embeddings

#NLP --

# ANN -- Aritifical Neural network -- text based -- huge amount of text
#!pip install tensorflow

#ANN -- Aritifical Neural network -- text based -- huge amount of text

import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense   # hidden layer

# training data
# Years of Exp
X=np.array([1,2,3,4,5],dtype=float)
# salary in thousands
y = np.array([30,35,40,45,50],dtype=float)

# build an ANN model
model = Sequential([
        Dense(10, activation='relu',input_shape=[1]),  # Dense -- hidden layer, input_Shape =no of parameters,
        Dense(10, activation='relu'),
        Dense(1)   # output -- salary
])

model.compile(optimizer='adam', loss='mse',metrics=['mae'])
#(optimizer---- equal weightage to all parameters,
# metrics-- to check the accuracy
# loss -- information lost

model.fit(X,y, epochs=500, verbose=0)
#epochs/runs/loop
#verbose=0  ---- hide the training logs

pred = model.predict(np.array([6]))
print("Predicted sal is:", pred[0][0])

loss, mae = model.evaluate(X,y, verbose=0)
print("Accuracy is", mae)
print("loss is ", loss)

model.save("salary_pred.keras")
print("model saved")
```
